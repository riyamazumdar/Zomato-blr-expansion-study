```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv('data/zomato.csv',thousands=',')
print("Shape:",df.shape)
df.head()
```

```python
print(df.info)
print("\nNull values:\n", df.isnull().sum())
print("\nColumns names:", df.columns.tolist())
```

```python
df.columns=[
    'url', 'address', 'name','online_order', 'book_table',
    'rate', 'votes', 'phone', 'location', 'rest_type', 
    'dish_liked', 'cuisines','cost_for_two', 'reviews_list',
    'menu_item', 'listed_type', 'listed_city'
]
```

```python
df.head()
```

```python
df['rate']=df['rate'].astype(str).str.replace('/5','', regex=False).str.strip()
df['rate']=df['rate'].replace(['NEW','-', 'nan'],np.nan)
df['rate']=pd.to_numeric(df['rate'], errors='coerce')

print("Rating range:", df['rate'].min(),"to", df['rate'].max())
print("Nulls in rate:", df['rate'].isnull().sum())
```

```python
print("Restaurants with ratings:", df['rate'].notna().sum())
print("Restaurants with ratings:", df['rate'].isna().sum())

```

```python
df.head()
```

```python
df['cost_for_two']=df['cost_for_two'].astype(str).str.replace(',','', regex=False)
df['cost_for_two']=pd.to_numeric(df['cost_for_two'], errors='coerce')

print("Cost range: Rs.", df['cost_for_two'].min(), "to Rs.", df['cost_for_two'].max())
print("Nulls in cost:", df['cost_for_two'].isnull().sum())
```

```python
median_cost=df['cost_for_two'].median()
df['cost_for_two'].fillna(median_cost, inplace=True)

print("Nulls after fill:", df['cost_for_two'].isnull().sum())
print("Median cost used: Rs.", median_cost)
```

```python
df.head(5)
```

```python
df['online_order']=df['online_order'].map({'Yes':1,'No':0})
df['book_table']=df['book_table'].map({'Yes':1,'No':0})
df['votes']=pd.to_numeric(df['votes'], errors='coerce').fillna(0).astype(int)

print("Online order values:", df['online_order'].unique())
print("Book table values:", df['book_table'].unique())
print("Votes range:",df['votes'].min(), "to", df['votes'].max())
```

```python
df.dropna(subset=['name', 'location'], inplace= True)
print("Shape after dropping essential nulls:", df.shape)
```

```python
df['cuisine_count']=df['cuisines'].dropna().apply(lambda x: len(x.split(',')))
df['primary_cuisine']=df['cuisines'].dropna().apply(lambda x: x.split(',')[0].strip())

df[['name','cuisines','primary_cuisine','cuisine_count']].head(10)
```
